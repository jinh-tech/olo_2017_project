```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lfm1 = np.loadtxt("ndcg_scores/cascade_ucb1/score_lfm_1.txt")
lfm2 = np.loadtxt("ndcg_scores/cascade_kl_ucb/score_lfm_1.txt")
lfm3 = np.loadtxt("ndcg_scores/dcm_kl_ucb/score_lfm_1.txt")
lfm4 = np.loadtxt("ndcg_scores/rba_klucb/score_lfm_1.txt")
lfm5 = np.loadtxt("ndcg_scores/rba_ucb/score_lfm_1.txt")
print "Graph for K=1 , Different Instances , Last FM "
plt.figure(figsize=(12,6))
plt.plot(lfm1,"red",label="cascade_ucb")
plt.plot(lfm2,"blue",label="cascade_klucb")
plt.plot(lfm3,"green",label="dcm_kl_ucb")
plt.plot(lfm4,"black",label="rba_klucb")
plt.plot(lfm5,"orange",label="rba_ucb")
plt.xlabel("Timesteps")
plt.ylabel("Average Cumulative NDCG Score")
plt.legend()
plt.grid()
plt.show()
```
